```python
def normalize_rgb_image(image):
  return image/255.
```

# affectnet-training-data Dataset

```python
dataset = os.path.join(DATA_FOLDER, 'affectnet-training-data')
```

```python
test_img_path = os.path.join(dataset,'anger','image0000006.jpg')

test_img = cv2.imread(test_img_path)
bw_img = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.imshow(test_img)
ax2.imshow(bw_img)

plt.show()
```

```python
df = pd.read_csv(os.path.join(dataset, 'labels.csv'))
df.head()
```

```python
labels = df['label'].unique().tolist()
```

```python
data_paths = { label:os.path.join(OUTPUT_IMAGES, label) for label in labels }
```

```python
for path in data_paths.values():
  print(f'mkdir: {path}')
  os.mkdir(path)
```

```python
for label in labels:
  print(label)
  for img_path in tqdm(df.loc[df['label'] == label]['pth']):
    image_name = img_path.split('/')[-1]
    img = cv2.imread(os.path.join(dataset, img_path))
    bw_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    normalized = normalize_rgb_image(bw_img)
    cv2.imwrite(os.path.join(data_paths[label], image_name), normalized)
```
